```python
def extract(year):

    date_list = []
    price_list = []

    for i in range(0,100,10):
        driver.get(f'https://www.amazon.ca/gp/your-account/order-history/ref=ppx_yo_dt_b_pagination_1_2?ie=UTF8&orderFilter=year-{year}&search=&startIndex={i}')

        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        time.sleep(1)

        info = soup.find_all('span', class_='a-color-secondary value')
        if len(info) > 0: # if it has tags proceed to extract, otherwise break
            dates = info[0::3]
            prices = info[1::3]

            for date in dates:
                date_list.append(date.text.strip())

            for price in prices:
                price_list.append(price.text.strip().replace('CDN$ ', ''))
        else:
            break

    df = pd.DataFrame(list(zip(date_list, price_list)))
    df['year'] = year

    return df
```

```python
df_2022 = extract(2022)
```

```python
df_2021 = extract(2021)
df_2021
```

```python
df_2020 = extract(2020)
df_2019 = extract(2019)
df_2018 = extract(2018)
```
